<b> Print out percentages of the ranges that are calculated based upon the generated data and Mean Square Error (MSE) between the original percentage and the generated data's percentage. We can see that the MSE is around E^-8, which are fairly small to prove the faithfulness to the original dataset.

```python
count = 0
def srange(s):
    range_age = s.split('_')
    floor = int(range_age[0])
    ceiling = int(range_age[1])
    return floor, ceiling
```

```python
for k in range(8):
    print("--------")
    column_ = list()
    floor_a, ceiling_a = srange(age_arr[k])
    for j in range(14):
        c = 0
        for index, i in df_1.iterrows():
            if (i["Age"] >= floor_a and i["Age"] <= ceiling_a):
                floor, ceiling = srange(sal_arr[j])
                if (i["Balance"] >= floor and i["Balance"] <= ceiling):
                    c += 1
        column_.append(c/1234)
        print(c/1234)
    print("mean square error (MSE): ") 
    cur = df_sal_age.iloc[0:-1, k+1].to_numpy()
    err = mean_squared_error(cur, column_)
    print(err)
```
